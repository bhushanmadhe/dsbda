```python
# Define features (based on your column names in df) 
features = ['SepalLength', 'SepalWidth', 'PetalLength', 'PetalWidth']
# Create subplots (2 rows, 2 columns) 
plt.figure(figsize=(12, 10)) 
# Loop through each feature to create a histogram 
for i, feature in enumerate(features, 1): 
   plt.subplot(2, 2, i)  # 2x2 grid, i-th subplot 
   plt.hist(df[feature], bins=20, color='skyblue', edgecolor='black') 
   plt.title(f'Histogram of {feature}') 
   plt.xlabel(feature) 
   plt.ylabel('Frequency') 
# Show all histograms 
plt.tight_layout() 
plt.show() 
```

```python
import seaborn as sns 
# Make sure your features list is defined correctly 
features = ['SepalLength', 'SepalWidth', 'PetalLength', 'PetalWidth'] 
# Create subplots (2 rows, 2 columns) 
plt.figure(figsize=(12, 8)) 
# Loop through each feature to create a boxplot 
for i, feature in enumerate(features, 1): 
  plt.subplot(2, 2, i)  # 2x2 grid 
  sns.boxplot(data=df, x=feature, color='lightgreen') 
  plt.title(f'Boxplot of {feature}') 
  plt.xlabel(feature) 
# Show all boxplots 
plt.tight_layout() 
plt.show() 
```

```python
# Create subplots: 4 rows, 2 columns (1 histogram and 1 boxplot per feature) 
plt.figure(figsize=(14, 20)) 
for i, feature in enumerate(features): 
# Histogram 
   plt.subplot(4, 2, 2*i + 1) 
   plt.hist(df[feature], bins=20, color='skyblue', edgecolor='black') 
   plt.title(f'Histogram of {feature}') 
   plt.xlabel(feature) 
   plt.ylabel('Frequency')
   # Boxplot 
   plt.subplot(4, 2, 2*i + 2) 
   sns.boxplot(data=df, x=feature, color='lightgreen') 
   plt.title(f'Boxplot of {feature}') 
   plt.xlabel(feature) 
   plt.tight_layout() 
   plt.show() 

```
